```python
import re
import numpy as np
import pandas as pd

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')

from nltk.stem.snowball import SnowballStemmer
spanishStemmer=SnowballStemmer("spanish")

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_distances
```

# Punto 1: Pre-Procesamiento

- `[9 pts]` Leer el archivo `bob_esponja.csv` usando `pandas`
- `[9 pts]` Crear una nueva columna con el texto en minúscula, sin caracteres especiales ni números, sin palabras vacias y hacer stemming de las palabras

```python
# Leer archivo
Bob_esponja = pd.read_csv ('bob_esponja.csv')
```

```python
print(Bob_esponja)
```

```python
# Nueva columna

def p_texs(tex):
    tex = tex.lower()
    tex = re.sub(r"[\W\d_]+", " ", tex)
    tex = " ".join([palabra for palabra in tex.split() if palabra not in stopwords_sp])
    tex = " ".join([spanishStemmer.stem(palabra) for palabra in tex.split()])
    return tex
```

```python
Bob_esponja['pre-procesado'] = Bob_esponja['Descripción'].apply(lambda tex: p_texs(tex))
Bob_esponja
```

# Punto 2: TF-IDF

- `[16 pts]` Crear la matriz TF-IDF

```python
# tfidf_matrix
matriz_vector = TfidfVectorizer()
tfid_bob = matriz_vector.fit_transform(Bob_esponja['pre-procesado'].values)
tfid_bob = pd.DataFrame(tfid_bob.toarray(), columns=matriz_vector.get_feature_names())
tfid_bob.index = Bob_esponja['Personaje'].values
tfid_bob.T.round(3)
```

# Punto 3: Distancia del coseno
- `[12 pts]` Calcular la distancia del coseno entre cada una de los personajes
- `[2 pts]` ¿Cuáles son los personajes más parecidos?
- `[2 pts]` ¿Cuáles son los personajes más diferentes?

```python
# dist_cos
dist_cos = cosine_distances(tfid_bob.values)
dist_cos = pd.DataFrame(dist_cos, columns = tfid_bob.T.columns, index = tfid_bob.T.columns)
dist_cos
```
